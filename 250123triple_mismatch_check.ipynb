```python
import pandas as pd

# 文件路径
file_path = '/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/data/241202spatial_join_2010.xlsx'

# 读取Excel文件
df = pd.read_excel(file_path)

# 定义筛选条件
def filter_rows(row):
    cntygb_1982 = row['CNTYGB_1982']
    cntygb_1990a = row['CNTYGB_1990a']
    cntygb_2000 = row['CNTYGB_2000']
    cntygb_2010 = row['CNTYGB_2010']

    # 非空值列表
    non_empty = [x for x in [cntygb_1982, cntygb_1990a, cntygb_2000] if pd.notna(x)]

    # 条件1：有一个非空，且和CNTYGB_2010不一致
    if len(non_empty) == 1 and non_empty[0] != cntygb_2010:
        return True

    # 条件2：两个非空，且两个中没有任意一个和CNTYGB_2010一样
    if len(non_empty) == 2 and all(x != cntygb_2010 for x in non_empty):
        return True

    # 条件3：三个非空，其中有两个是一样的，但是这个一样的和CNTYGB_2010不一样
    if len(non_empty) == 3:
        unique_values = set(non_empty)
        if len(unique_values) == 2 and cntygb_2010 not in unique_values:
            return True

    # 条件4：三个非空，三个都不一样，且没有任何一个和CNTYGB_2010一样
    if len(non_empty) == 3:
        unique_values = set(non_empty)
        if len(unique_values) == 3 and all(x != cntygb_2010 for x in unique_values):
            return True

    return False

# 应用筛选条件
filtered_df = df[df.apply(filter_rows, axis=1)]
```
