```python
import pandas as pd
import re

# 示例数据（请用你的实际数据替换）
# df = pd.DataFrame({
#     'name': ['张三', '李四', '王五', '赵六'],
#     'area': ['东部苗族聚居区', '西部汉族区域', '南部回族社区', '北部壮族村落']
# })
# minority = pd.DataFrame({
#     'name': ['苗族', '回族', '壮族'],
#     'id': [101, 102, 103]
# })

# 1. 筛选出 df 中 area 列含有 “族” 的行，同时排除 area 中含有 “汉族” 的行
filtered_df = df[
    df['area'].str.contains('族', na=False) & ~df['area'].str.contains('汉族', na=False)
].copy()

# 2. 构造正则表达式模式，将 minority 表中所有少数民族名称拼接成形如 '苗族|回族|壮族' 的模式
pattern = '|'.join(re.escape(item) for item in minority['name'].dropna().unique())

# 3. 在 filtered_df 的 area 列中提取匹配到的少数民族名称，赋值到新列 'Ethnicity'
filtered_df.loc[:, 'Ethnicity'] = filtered_df['area'].str.extract('(' + pattern + ')', expand=False)

# 4. 构造映射字典，从 minority 表中将民族名称映射到对应的 id
ethnic_mapping = minority.set_index('name')['id']

# 5. 根据提取到的 Ethnicity 得到对应的 Ethnic_ID，新建一列 'Ethnic_ID'
filtered_df.loc[:, 'Ethnic_ID'] = filtered_df['Ethnicity'].map(ethnic_mapping)

# 6. 构造最终新表，只保留需要的列：name、area、Ethnicity、Ethnic_ID
new_df = filtered_df[['name', 'area', 'Ethnicity', 'Ethnic_ID']]

new_df
```
