```python
bert_similarity_scores_df = pd.merge(
    bert_similarity_scores_df,
    spatial_join_df[['name', 'CNTYGB_2010']],  # 只保留需要的列
    on='name',  # 按照 'name' 列合并
    how='left'  # 使用左连接，保留 bert_similarity_scores_df 的所有行
)
bert_similarity_scores_df.head()
```

```python
import pandas as pd
# 筛选列名包含“相似度”的列
similarity_columns = [col for col in bert_similarity_scores_df.columns if '相似度' in col]
# 按 CNTYGB_2010_y 分组并计算包含“相似度”列的平均值
grouped_mean = bert_similarity_scores_df.groupby('CNTYGB_2010_y')[similarity_columns].mean()
grouped_mean
```

```python
# 替换列名为统一格式
grouped_mean.rename(
    columns=lambda x: f"bert_similarity_{x.split('_')[0]}" if '_相似度' in x else x,
    inplace=True
)
grouped_mean
```

```python
grouped_mean.reset_index(inplace=True)
grouped_mean
```

```python
grouped_mean.reset_index(inplace=True)
if 'CNTYGB_2010_y' in grouped_mean.columns:
    # 将CNTYGB_2010_y列转换为整数类型，处理NaN值为默认值（如0）
    grouped_mean['CNTYGB_2010_y'] = grouped_mean['CNTYGB_2010_y'].fillna(0).astype(int)
else:
    print("列名 'CNTYGB_2010_y' 不存在，请检查数据框")
grouped_mean
```

```python
# 确保列名存在
if 'china10_cntygb' in cgss_data_df.columns:
    # 将 china10_cntygb 列转换为整数类型，空值填充为 0
    cgss_data_df['china10_cntygb'] = cgss_data_df['china10_cntygb'].fillna(0).astype(int)
else:
    print("列名 'china10_cntygb' 不存在，请检查数据框")
cgss_data_df.head()
```

```python
# 确保需要合并的列存在
if 'CNTYGB_2010_y' in grouped_mean.columns and 'china10_cntygb' in cgss_data_df.columns:
    # 按照 CNTYGB_2010_y 和 china10_cntygb 列的值进行合并
    merged_df = cgss_data_df.merge(grouped_mean, left_on='china10_cntygb', right_on='CNTYGB_2010_y', how='left')
    # 删除 CNTYGB_2010_y 列
    merged_df.drop(columns=['CNTYGB_2010_y'], inplace=True)
merged_df.head()
```

```python
import pandas as pd

# 假设数据框名为 df
# 筛选主题列
theme_columns = spatial_join_df.columns[2:]  # 从第4列开始是主题列

# 创建一个空的结果数据框
results = []

# 按照 CNTYGB_2010 分组
grouped = spatial_join_df.groupby('CNTYGB_2010')

# 遍历每个分组
for cntygb, group in grouped:
    # 对每个主题统计 A, B, C, D 的占比
    for theme in theme_columns:
        counts = group[theme].value_counts(normalize=True)  # 计算占比
        for option in ['A', 'B', 'C', 'D']:
            results.append({
                'CNTYGB_2010': cntygb,
                'Theme': theme,
                'Option': option,
                'Proportion': counts.get(option, 0)  # 若没有该选项则填0
            })

# 转换结果为 DataFrame
proportion_df = pd.DataFrame(results)
proportion_df.head()
```

```python
# 将列名组合为 Theme+Option
proportion_df['Theme_Option'] = proportion_df['Theme'] + '_' + proportion_df['Option']

# 转换为宽格式
wide_df = proportion_df.pivot(
    index='CNTYGB_2010',  # 行索引为 CNTYGB_2010
    columns='Theme_Option',  # 列索引为 Theme_Option
    values='Proportion'  # 值为 Proportion
)

# 重置列索引，恢复为普通列
wide_df.reset_index(inplace=True)

wide_df
```

```python
wide_df.columns = [
    col if i < 1 else f"gpt_proportion_{col}" 
    for i, col in enumerate(wide_df.columns)
]
wide_df
```

```python
wide_df['CNTYGB_2010'] = wide_df['CNTYGB_2010'].fillna(0).astype(int)
```
