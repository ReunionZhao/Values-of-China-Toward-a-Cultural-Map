1. tale-level的bert similarity的correlation

```python
import pandas as pd

data0215 = pd.read_csv("/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250307data_to_check/250215similarity_scores_5_versions/250215bert_similarity_scores_version_1.csv")
data0222 = pd.read_csv("/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250307data_to_check/250222similarity_scores_5_versions/250222bert_similarity_scores_version_1.csv")
```

```python
# 给所有包含 "相似度" 的列名加上 0215 或 0222 后缀
data0215 = data0215.rename(columns={col: col + "_0215" for col in data0215.columns if "相似度" in col})
data0222 = data0222.rename(columns={col: col + "_0222" for col in data0222.columns if "相似度" in col})

# 依据 name 列合并两个表
merged_data = pd.merge(data0215, data0222, on="name", how="outer")

# 获取所有包含 "相似度" 的列名
similarity_cols_0215 = {col.replace("_0215", "") for col in data0215.columns if "相似度" in col}
similarity_cols_0222 = {col.replace("_0222", "") for col in data0222.columns if "相似度" in col}

# 找到两个表都含有的相似度列
common_similarity_cols = similarity_cols_0215 & similarity_cols_0222

# 输出结果
print("相同的相似度列（去除后缀）：", list(common_similarity_cols))
```

```python
from scipy.stats import spearmanr

# 创建一个存储相关性结果的列表
correlation_results = []

# 遍历共同的相似度列
for col in common_similarity_cols:
    col_0215 = col + "_0215"
    col_0222 = col + "_0222"
    
    valid_data = merged_data[[col_0215, col_0222]].dropna()
    
    if not valid_data.empty:
        # 计算 Spearman 相关系数
        corr, p_value = spearmanr(valid_data[col_0215], valid_data[col_0222])
        
        # 相关系数保留 7 位小数
        corr = format(corr, ".7f")
        
        # p 值转换为科学计数法，并保留 7 位小数
        p_value = format(p_value, ".7e")
        
        correlation_results.append([col, corr, p_value])

correlation_df = pd.DataFrame(correlation_results, columns=["相似度列", "Spearman相关系数", "p值"])

correlation_df.to_excel("/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250307data_to_check/250307correlation_version1.xlsx")
```
