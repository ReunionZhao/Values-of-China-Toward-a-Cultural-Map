读取少数民族故事名称、并抽取故事

（把少数民族故事去除后再找40个，风格会更统一一点）

````python
import os
import random
import pandas as pd
import csv

# 加载包含少数民族故事名称的文件
file_path = "/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/data/Merged_Ethnicity_Data_with_English_Names_Xue.csv"
shaoshuminzu = pd.read_csv(file_path)

# 获取少数民族故事名称的列表
minority_story_names = set(shaoshuminzu['name'])

# 指定主文件夹路径
main_folder_path = '/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/data/raw_data_tales'

# 输出 CSV 文件路径
output_csv_path = '/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250107sample_stories_without_minority.csv'

def load_and_save_stories(folder_path, excluded_names, output_csv_path, sample_size=30):
    stories = []

    # 遍历主文件夹下的所有子文件夹及文件
    for root, _, files in os.walk(folder_path):
        for filename in files:
            # 获取文件名，不包含扩展名
            name_without_extension = os.path.splitext(filename)[0]

            # 如果文件名（不含扩展名）在排除名单中，跳过
            if name_without_extension in excluded_names:
                continue
            
            file_path = os.path.join(root, filename)

            # 仅处理文本文件
            if filename.endswith('.txt'):
                with open(file_path, 'r', encoding='utf-8') as file:
                    story_content = file.read()
                    stories.append({'Tale Name': name_without_extension, 'Tale Content': story_content})
    
    # 随机抽取指定数量的故事
    sampled_stories = random.sample(stories, min(sample_size, len(stories)))
    
    # 保存抽取的故事到 CSV 文件
    with open(output_csv_path, 'w', encoding='utf-8', newline='') as csvfile:
        fieldnames = ['Tale Name', 'Tale Content']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        writer.writerows(sampled_stories)
    
    print(f"抽取的故事已保存到 {output_csv_path}")
    return sampled_stories

# 抽取并保存故事
sampled_stories = load_and_save_stories(main_folder_path, minority_story_names, output_csv_path)
````

生成风格摘要

````python
import openai
import csv
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor, as_completed

# 从每个完整的故事生成风格描述
def extract_style_segment(tale_name, tale_content):
    prompt = f"你是一位精通中国民间故事研究的专家，请从叙述方式、语言风格和结构特点三个维度，分析并提炼其表达特征与创作手法。注意，需要对以下故事进行高度概括的分析，不要涉及任何故事的具体内容（比如人物、地点或情节等），答案需简洁精准，无需补充背景或细节说明，字数不超过200字。以下是故事内容：{tale_content}"    
    try:
        response = client.chat.completions.create(
            model="gpt-4o",  # 请根据需要调整模型名称
            messages=[{"role": "user", "content": prompt}]
        )
        style_summary = response.choices[0].message.content.strip()
        #print(tale_name)
        #print(style_summary)
        return tale_name, tale_content, style_summary
    except Exception as e:
        print(f"Error processing text segment: {e}")
        return tale_name, tale_content, "Error"

# 并行处理所有故事并保存到 CSV 文件
def generate_and_save_style_summaries(stories):
    output_file = "/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250107story_styles_summary.csv"
    
    with open(output_file, mode="w", newline='', encoding="utf-8-sig") as file:
        writer = csv.writer(file)
        writer.writerow(["故事名称", "故事内容", "故事风格摘要"])
        
        with ThreadPoolExecutor(max_workers=25) as executor:
            # 提交每个故事的风格描述生成任务到线程池
            futures = {executor.submit(extract_style_segment, story['Tale Name'], story['Tale Content']): story for story in stories}
            
            # 处理完成的任务并写入 CSV 文件
            for future in tqdm(as_completed(futures), total=len(futures), desc="Processing Stories"):
                tale_name, tale_content, style_summary = future.result()
                writer.writerow([tale_name, tale_content, style_summary])

    print(f"所有故事的风格摘要已保存到 {output_file}")

# 执行风格摘要生成并保存
generate_and_save_style_summaries(sampled_stories)
````

````python
def generate_story_with_json_conversion(style_summary, theme, explanation, reference_stories):
    # 构造参考故事部分
    references = "\n\n".join([f"参考故事{i+1}：{story}" for i, story in enumerate(reference_stories)])
    
     # 构造完整的提示语
    prompt = f"""
    你是一位擅长撰写中国民间故事的作家。

    需要你生成一个主题为“{theme}”的中国民间故事。具体而言：{explanation}

    请注意：这是一个围绕单一主题的故事，不要同时包含其他主题。

    在写作时，请参考以下信息：
    1. 故事风格特点摘要：{style_summary}
    2. 五个参考故事：{references}
    具体要求如下：
    1. 在语言和叙述方式等写作风格方面，请总结并遵循故事风格摘要与五个参考故事的特点。
    2. 在时代背景方面，请与五个参考故事保持一致。

    请将以上要求自然融入故事中，字数控制在1000字左右，并只输出生成的故事。

    此外，请写一小段说明，解释为何这个故事符合上述主题“{theme}”“{explanation}”的要求。

    输出内容请严格遵循以下格式，不需要任何其余提示词：
    ```
    {{
      "theme": "{theme}",
      "story": "生成的故事内容",
      "justification": "说明故事与主题的相关性"
    }}
    ```
    """

    
    try:
        # 第一阶段：使用 o1-mini 生成伪 JSON 格式的字符串
        response = client.chat.completions.create(
            model="o1-mini",
            messages=[{"role": "user", "content": prompt}]
        )
        json_string = response.choices[0].message.content.strip()

        # 构造第二阶段提示语
        conversion_prompt = f"""
        以下是一个伪JSON格式的字符串。它包含三个字段：1）theme：故事的主题。2）story：生成的故事内容。3）justification：说明故事与主题的相关性。
        
        你的任务是：1）确保将其解析为标准的 JSON 格式。2）确保三个字段的内容保持原样，不要修改任何内容。
        
        以下是需要解析的字符串：
        {json_string}
        
        请将其转换为有效的 JSON 格式并返回。
        """
        
        # 第二阶段：使用 gpt-4o 处理为 JSON 格式
        response_json = client.chat.completions.create(
            model="gpt-4o",
            messages=[{"role": "user", "content": conversion_prompt}],
            response_format={"type": "json_object"}
        )
        result = response_json.choices[0].message.content.strip()

        # 确保返回有效 JSON
        try:
            result_json = json.loads(result)
            return json.dumps(result_json, ensure_ascii=False, indent=4)
        except json.JSONDecodeError as e:
            print(f"JSON 解析错误: {e}")
            return json.dumps({"error": "Error generating JSON format"}, ensure_ascii=False, indent=4)

    except Exception as e:
        print(f"Error generating story for {theme}: {e}")
        return json.dumps({"error": "Error generating story"}, ensure_ascii=False, indent=4)
````

````python
# 定义处理每一行的函数
def process_row(row, style_summary, reference_stories):
    theme = row['Theme']  # 获取主题名称

    results = []
    # 遍历从第二列到最后一列，逐一传递给generate_story函数
    for column in ['Generic', 'Specific', 'Specific_正面', 'Specific_反面', 'Specific_中立']:
        explanation = row[column]
        if pd.notna(explanation):  # 如果该列有内容
            story_json = generate_story_with_json_conversion(style_summary, theme, explanation, reference_stories)
            print(f"Generated story JSON for theme '{theme}' and column '{column}': {story_json}")
            story_data = json.loads(story_json)
            
            # 组合数据并保存
            result = {
                'Theme': theme,
                'Explanation': explanation,
                'Explanation_Column': column,  # 保存当前列名
                'Story': story_data.get('story', ''),
                'Justification': story_data.get('justification', '')
            }
            results.append(result)
    
    return results

# 定义并行处理的主函数，使用tqdm显示进度条
def process_rows_parallel(df, style_summary, reference_stories):
    all_results = []
    
    # 使用ThreadPoolExecutor进行并行处理，并用tqdm显示进度条
    with ThreadPoolExecutor(max_workers=45) as executor:  # 设置最大并行工作线程数为45
        # 提交任务，并将任务与进度条结合
        futures = [executor.submit(process_row, row, style_summary, reference_stories) for index, row in df.iterrows()]
        
        # 使用tqdm显示进度
        for future in tqdm(futures, desc="Processing rows", total=len(futures)):
            result = future.result()
            all_results.extend(result)
    
    return all_results

file_path = '/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/data/250107latest_prompts_for_synthetic_archetypes.xlsx'
df = pd.read_excel(file_path)

# 调用并行处理函数
results = process_rows_parallel(df, style_summary, reference_stories)

# 将结果保存到新的 CSV 文件
output_file_path = '/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250107generated_synthetic_story_distinguished_attitude.xlsx'
results_df = pd.DataFrame(results)
results_df.to_excel(output_file_path, index=False)

print(f"故事和解释已成功保存到 {output_file_path}")
````
