```python
import pandas as pd

# Define base file paths
bert_base_path = "/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250222similarity_scores_5_versions/250222bert_similarity_scores_version_{}.csv"
gpt_file_path = "/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250208tales_final_cgss_county_level/250208gpt_advanced_multitopic_story_analysis.xlsx"
output_base_path = "/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250222tale-level_score_bert+gpt_version{}.xlsx"

# Load the GPT analysis file (constant for all merges)
gpt = pd.read_excel(gpt_file_path)

# Iterate over versions 1 to 5
for version in range(1, 6):
    # Load the BERT similarity scores for the current version
    bert_file_path = bert_base_path.format(version)
    bert = pd.read_csv(bert_file_path)

    # Merge BERT and GPT data
    merged_df = bert.merge(gpt, left_on='name', right_on='故事名称', how='inner')

    # Reorder columns to move "故事原文" to the last position
    column_order = [col for col in merged_df.columns if col != '故事原文'] + ['故事原文']
    merged_df = merged_df[column_order]

    # Drop unnecessary columns if they exist
    columns_to_drop = ["Unnamed: 0.1", "Unnamed: 0", "故事名称"]
    merged_df = merged_df.drop(columns=[col for col in columns_to_drop if col in merged_df.columns])

    # Save the merged result to an Excel file
    output_file_path = output_base_path.format(version)
    merged_df.to_excel(output_file_path, index=False)
    print(f"✅ Saved: {output_file_path}")
```

```python
import pandas as pd

merged_df = pd.read_excel('/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250222tale-level_score_bert+gpt_version5.xlsx')

# 去除少数民族
filtered_merged_df = merged_df[~merged_df['name'].isin(df['name'])]
# If you want to overwrite the original merged_df
merged_df = filtered_merged_df.reset_index(drop=True)


# 筛选列名包含“相似度”的列
similarity_columns = [col for col in merged_df.columns if '相似度' in col]
# 按 CNTYGB_2010_y 分组并计算包含“相似度”列的平均值
grouped_mean = merged_df.groupby('CNTYGB_2010')[similarity_columns].mean()
# 替换列名为统一格式，去掉结尾的“_相似度”
grouped_mean.rename(
    columns=lambda x: f"bert_similarity_{x.rsplit('_相似度', 1)[0]}" if x.endswith('_相似度') else x,
    inplace=True
)
grouped_mean.reset_index(inplace=True)
if 'CNTYGB_2010' in grouped_mean.columns:
    # 将CNTYGB_2010_y列转换为整数类型，处理NaN值为默认值（如0）
    grouped_mean['CNTYGB_2010'] = grouped_mean['CNTYGB_2010'].fillna(0).astype(int)
else:
    print("列名 'CNTYGB_2010' 不存在，请检查数据框")
# 确保列名存在
if 'china10_cntygb' in cgss_data_df.columns:
    # 将 china10_cntygb 列转换为整数类型，空值填充为 0
    cgss_data_df['china10_cntygb'] = cgss_data_df['china10_cntygb'].fillna(0).astype(int)
else:
    print("列名 'china10_cntygb' 不存在，请检查数据框")
# 确保需要合并的列存在
if 'CNTYGB_2010' in grouped_mean.columns and 'china10_cntygb' in cgss_data_df.columns:
    # 按照 CNTYGB_2010_y 和 china10_cntygb 列的值进行合并
    merged_df2 = cgss_data_df.merge(grouped_mean, left_on='china10_cntygb', right_on='CNTYGB_2010', how='left')
    # 删除 CNTYGB_2010_y 列
    merged_df2.drop(columns=['CNTYGB_2010'], inplace=True)

import re

# Function to modify column names
def clean_bert_column_names(col_name):
    if 'bert' in col_name:
        # Remove Chinese characters
        col_name = re.sub(r'[\u4e00-\u9fff]+', '', col_name)
        # Remove "_similarity" and anything before it
        col_name = re.sub(r'.*_similarity', '', col_name)
        # Remove parentheses (both Chinese and English)
        col_name = col_name.replace('（', '').replace('）', '').replace('(', '').replace(')', '')
        # Add "bert_" prefix
        col_name = f"bert{col_name}"
    return col_name

# Apply function to rename columns
merged_df2.columns = [clean_bert_column_names(col) for col in merged_df2.columns]

merged_df2.columns = merged_df2.columns.str.replace("__", "_", regex=False)

# 对 merged_df2 进行重命名（确保 merged_df2 已是 DataFrame）
merged_df2.rename(columns=rename_mapping, inplace=True)

# 定义保存路径
output_csv_path = '/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/250222cgss_bert_county_level_mean_version_5.csv'
# 保存为 CSV 文件
merged_df2.to_csv(output_csv_path, index=False, encoding='utf-8-sig')
print(f"合并后的数据已保存到: {output_csv_path}")
```
