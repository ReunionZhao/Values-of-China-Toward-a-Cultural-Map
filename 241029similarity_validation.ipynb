```python
import pandas as pd

file_path = '/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/241028bert+gpt_tales_topic_similarity.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 筛选出包含“诚实”关键词的行和不包含“诚实”关键词的行
contains_honesty = data[data['内容'].str.contains('诚实', na=False)]
no_honesty = data[~data['内容'].str.contains('诚实', na=False)]

# 获取“与伪故事相似度”列数据
similarity_with_honesty = contains_honesty['与伪故事相似度']
similarity_without_honesty = no_honesty['与伪故事相似度']
```

```python
# 计算分位数和均值
quantiles_with_honesty = contains_honesty['与伪故事相似度'].quantile([0, 0.25, 0.5, 0.75, 1])
quantiles_without_honesty = no_honesty['与伪故事相似度'].quantile([0, 0.25, 0.5, 0.75, 1])
mean_with_honesty = contains_honesty['与伪故事相似度'].mean()
mean_without_honesty = no_honesty['与伪故事相似度'].mean()

# 输出结果
print("包含“诚实”的分位数：")
print(quantiles_with_honesty)
print("均值:", mean_with_honesty)

print("\n不包含“诚实”的分位数：")
print(quantiles_without_honesty)
print("均值:", mean_without_honesty)
```

```python
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(contains_honesty['与伪故事相似度'], no_honesty['与伪故事相似度'], equal_var=False)
print("\nT检验结果：")
print("t统计量:", t_stat)
print("p值:", p_value)
```

```python
file_path = '/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/data/raw_data_tales/地图匹配_0426_v4.csv'
base = pd.read_csv(file_path)

base.head()
```

```python
import re
# 将“与伪故事相似度”列与 base 表合并，依据 data 表的“文件名”和 base 表的“name”列
data['文件名'] = data['文件路径'].apply(lambda x: re.search(r'\/(.*?)\.txt', x).group(1) if re.search(r'\/(.*?)\.txt', x) else None)
merged_base = base.merge(data[['文件名', '与伪故事相似度']], left_on='name', right_on='文件名', how='left')
merged_base = merged_base.drop(columns=['文件名'])
merged_base.head()
```

```python
# 导出合并结果到 CSV 文件
output_path = '/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/241029bert+gpt_tales_topic_similarity_merged.csv'
merged_base.to_csv(output_path, index=False, encoding='utf-8-sig')
```
