```python
import pandas as pd
import numpy as np
from scipy import stats

columns = [
    '佛教_相似度', '诚实_相似度', '仁义_相似度', '儒家_相似度', '长寿_相似度', '努力_相似度', 
    '杀戮_相似度', '平等_相似度', '古时候_相似度', '忠诚_相似度', '结拜_相似度', '馈赠_相似度', 
    '道教_相似度', '宗族_相似度', '遵守_相似度', '帝王_相似度', '信任_相似度', '鬼神_相似度', 
    '慈善_相似度', '惩罚_相似度', '祖先_相似度', '孝道_相似度'
]

# 仅保留这些列
df1_filtered = df1[columns]
df2_filtered = df2[columns]

# 删除含有缺失值的行
df1_clean = df1_filtered.dropna()
df2_clean = df2_filtered.dropna()

# 存储结果的 DataFrames
desc_stats_df1 = pd.DataFrame()
desc_stats_df2 = pd.DataFrame()
t_test_results_df = pd.DataFrame()
correlation_results_df = pd.DataFrame()

# 计算描述性统计、t检验、相关性
for col in columns:
    # 描述性统计，添加分位数和样本个数
    desc_stats_df1[col] = df1_clean[col].describe(percentiles=[.25, .5, .75])  # 默认显示 25%, 50%, 75% 分位数
    desc_stats_df2[col] = df2_clean[col].describe(percentiles=[.25, .5, .75])  # 默认显示 25%, 50%, 75% 分位数
    
    # t检验
    t_stat, p_value = stats.ttest_ind(df1_clean[col], df2_clean[col], nan_policy='omit')  # 排除缺失值
    t_stat_label = f"{t_stat:.4f}"  # 保留 4 位小数
    p_value_label = f"{p_value:.4e}"  # 使用科学记数法表示 p 值，保留 2 位有效数字
    if p_value < 0.001:
        t_stat_label += " ***"  # p值小于0.001时标记 ***
    elif p_value < 0.01:
        t_stat_label += " **"  # p值小于0.01时标记 **
    elif p_value < 0.05:
        t_stat_label += " *"  # p值小于0.05时标记 *
    
    t_test_results_df[col] = [t_stat_label, p_value_label]
    
    # 相关性假设检验（Pearson）
    correlation, p_corr_value = stats.pearsonr(df1_clean[col], df2_clean[col])
    corr_label = f"{correlation:.4f}"  # 保留 4 位小数
    p_corr_label = f"{p_corr_value:.4e}"  # 使用科学记数法表示 p 值，保留 2 位有效数字
    if p_corr_value < 0.001:
        corr_label += " ***"  # p值小于0.001时标记 ***
    elif p_corr_value < 0.01:
        corr_label += " **"  # p值小于0.01时标记 **
    elif p_corr_value < 0.05:
        corr_label += " *"  # p值小于0.05时标记 *
    
    correlation_results_df[col] = [corr_label, p_corr_label]

# 为 t-test 和相关性结果添加行标签
t_test_results_df.index = ['t-statistic', 'p-value']
correlation_results_df.index = ['correlation', 'p-value']
```

```python
import pandas as pd

# 定义文件路径
file_path_1 = "/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/241107generated_stories.xlsx"
file_path_2 = "/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/241031generated_stories.csv"

# 读取文件并命名为 df1 和 df2
df1 = pd.read_excel(file_path_1)  # 读取 Excel 文件
df2 = pd.read_csv(file_path_2)    # 读取 CSV 文件
```

```python
df1.head()
```

```python
# 假设 df1 原来的列名是 ['old_col1', 'old_col2', 'old_col3']
new_column_names = {
    '目的': 'intention_目的',
    '生成的故事': 'intention_生成的故事'
}

# 使用 rename 修改列名
df1.rename(columns=new_column_names, inplace=True)
```

```python
df2.head()
```

```python
# 假设 df1 原来的列名是 ['old_col1', 'old_col2', 'old_col3']
new_column_names = {
    '释义': 'definition+topic_释义',
    '生成的故事': 'definition+topic_生成的故事'
}

# 使用 rename 修改列名
df2.rename(columns=new_column_names, inplace=True)
```

```python
merged_df = pd.merge(df1, df2, on='主题', how='left')  # 使用 left join 保留 df1 的所有行
merged_df
```

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer

# 加载模型
model = SentenceTransformer('DMetaSoul/sbert-chinese-general-v2')
# 定义计算相似度的函数
def compute_similarity(text1, text2):
    embeddings = model.encode([text1, text2])
    return np.dot(embeddings[0], embeddings[1]) / (
        np.linalg.norm(embeddings[0]) * np.linalg.norm(embeddings[1])
    )

# 计算相似度
merged_df["gpt_input_相似度"] = merged_df.apply(
    lambda row: compute_similarity(row["intention_目的"], row["definition+topic_释义"]), axis=1
)
merged_df["gpt_output_相似度"] = merged_df.apply(
    lambda row: compute_similarity(row["intention_生成的故事"], row["definition+topic_生成的故事"]), axis=1
)
```

```python
# 描述性统计
stats_df = merged_df[["gpt_input_相似度", "gpt_output_相似度"]].describe(percentiles=[.25, .5, .75])

# 保存到 Excel
output_path = "/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/result/241116generated_stories_similarity_results.xlsx"
with pd.ExcelWriter(output_path) as writer:
    merged_df.to_excel(writer, sheet_name="Merged Data", index=False)
    stats_df.to_excel(writer, sheet_name="Statistics")

print(f"数据和统计结果已保存到 {output_path}")
```
