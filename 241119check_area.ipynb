```python
import pandas as pd
# 文件路径
file_path = "/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/data/241119null_matched_areas_manual check.xlsx"

data = pd.read_excel(file_path)
filtered_data = data[(data["？"].notna()) & (data["？"] != "？")]
filtered_data.head()
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
from tqdm import tqdm

# 假设 filtered_data 和 df 已经加载

# 获取 df 的 "new_column" 列的值列表
new_column_values = df["new_column"].tolist()

# 定义一个函数，使用 fuzzywuzzy 模糊匹配，并返回最佳匹配结果和匹配度
def fuzzy_match(row, column_values, threshold=30):
    # 使用 process.extractOne 提取与“？”列值匹配度最高的结果
    best_match, score = process.extractOne(row["？"], column_values, scorer=fuzz.partial_ratio)
    
    # 如果匹配度高于阈值，返回最佳匹配和匹配度，否则返回 None 和 0
    if score >= threshold:
        return pd.Series([best_match, score])
    return pd.Series([None, 0])

# 使用 tqdm 显示进度条，在 filtered_data 上迭代并应用 fuzzy_match
tqdm.pandas(desc="Processing fuzzy matches")

# 对 filtered_data 中的每一行应用 fuzzy_match 函数，并将结果分配到新列
filtered_data[["county2000_matched_value", "county2000_match_score"]] = filtered_data.progress_apply(fuzzy_match, axis=1, column_values=new_column_values)

# 打印前几行查看匹配结果
print(filtered_data.head())

# 如果需要保存结果，可以用如下方式保存
# filtered_data.to_excel("/path/to/save/matched_data.xlsx", index=False)
```
