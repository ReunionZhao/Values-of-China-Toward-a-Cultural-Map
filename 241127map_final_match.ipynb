```python
def remove_unique_values(text, unique_values):
    if isinstance(text, str):  # 确保是字符串类型
        for value in unique_values:
            text = text.replace(value, '', 1)  # 只替换第一次出现的 value
    return text

# 应用这个函数到 matched_area 列
excel_data_manual['matched_area'] = excel_data_manual['matched_area'].apply(remove_unique_values, unique_values=unique_values)
```

```python
import pandas as pd
from openpyxl import load_workbook

# 读取第二个 Excel 文件
excel_file2 = '/Users/zhaorunping/Desktop/Research_Onging/2410_LSE_Xue/data/MAP241119match_area_gbcode.xlsx'

# 使用 openpyxl 读取 Excel 文件
wb = load_workbook(excel_file2, data_only=True)  # data_only=True 表示读取计算后的值，而不是公式
sheet = wb.active  # 默认读取第一个 sheet，你可以根据需要更改为其他 sheet

# 使用 pandas 读取数据
excel_data_third = pd.read_excel(excel_file2)

# 获取 "name" 列的索引
name_column_index = excel_data_third.columns.get_loc('name')

# 创建一个新的空列表，用于存储 "correct" 列的值
correct_column = []

# 遍历 "name" 列，判断每个单元格是否有底色
for row in range(2, sheet.max_row + 1):  # 从第2行开始，因为通常第1行是标题
    cell = sheet.cell(row=row, column=name_column_index + 1)  # openpyxl索引从1开始
    fill = cell.fill
    # 判断是否有底色，并将结果添加到列表
    if fill.start_color.index != '00000000':  # '00000000' 表示没有填充颜色
        correct_column.append(True)  # 有底色
    else:
        correct_column.append(False)  # 没有底色

# 将 "correct" 列添加到 DataFrame 中
excel_data_third['correct'] = correct_column
```

```python
# 1. 找到所有包含 'match_score' 的列名
score_columns = [col for col in df.columns if 'match_score' in col]

# 2. 新增列 'score' 为最大值
df['score'] = df[score_columns].max(axis=1)

# 3. 根据最大分数所在列的索引，找到相应的 'matched_value' 列
for idx, row in df.iterrows():
    max_score_col = row[score_columns].idxmax()  # 获取最大分数所在的列名
    matched_value_col = max_score_col.replace('match_score', 'matched_value')  # 找到对应的 'matched_value' 列名
    df.at[idx, 'county'] = row[matched_value_col]  # 设置 'county' 列为对应的 'matched_value'
```
